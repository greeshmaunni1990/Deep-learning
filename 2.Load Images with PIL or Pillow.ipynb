## Load and display images in matplotlib

```python
from matplotlib import image
import matplotlib.pyplot as plt

# imread will load the image in matplotlib
data = image.imread('open_house.jpg')

print(data.dtype)
print(data.shape)

# display the image inside matplotlib
plt.imshow(data)
plt.show()
```

## Load the image and convert the image to array and array back to image

```python
from PIL import Image
from numpy import asarray

# load the image using PIL
image = Image.open('open_house.jpg')

# convert image to numpy array
data = asarray(image)

print(data.shape)

# converting the array back to image
image2 = Image.fromarray(data)

print(image2.format)
print(image2.mode)
print(image2.size)
```

## Save image in other formats

```python
from PIL import Image

# load the image using PIL
image = Image.open('open_house.jpg')

# save the image in PNG format
image.save('open_house.png', format='PNG')

# load the image after converting to other formats
image2 = Image.open('open_house.png')

print(image2.format)
```

## Converting color image to greyscale

```python
# load the image using PIL
image = Image.open('open_house.jpg')

# convert the image to greyscale
greyscale_image = image.convert(mode='L')

# save the new image in .jpg format
greyscale_image.save('opera_house_geryscale.jpg')

# load the greyscale image usin PIL
image2 = Image.open('opera_house_geryscale.jpg')
image2.show()
```

## Resize an image with aspect ratio

```python
# Creating thumbnail of an image of aspect ratio (100,100)

# load the image
image = Image.open('open_house.jpg')
print(image.size)

# create a thumbnail
image.thumbnail((100, 100))

print(image.size)
image.show()
```

## Resize an image to new shape

```python
# load the image
image = Image.open('open_house.jpg')
print(image.size)

# resize will resize the image with the given width and height as the parameter
image_resized = image.resize((200,200))
print(image_resized.size)
image_resized.show()
```

## Data Augmentation(Flip, Rotate and Crop)

```python
# FLIP IMAGE


# load image
image = Image.open('open_house.jpg')

# horizontal flip
hor_flip = image.transpose(Image.FLIP_LEFT_RIGHT)

# vertical flip
vertical_flip = image.transpose(Image.FLIP_TOP_BOTTOM)

#plot the images
plt.figure(figsize=(20, 10))
plt.subplot(311)
plt.imshow(image)

plt.subplot(312)
plt.imshow(hor_flip)

plt.subplot(313)
plt.imshow(vertical_flip)
plt.show()
```

```python
# Rotate image

# load the image
image = Image.open('open_house.jpg')

plt.figure(figsize=(20, 10))
plt.subplot(311)
plt.imshow(image)

plt.subplot(312)
plt.imshow(image.rotate(45))

plt.subplot(313)
plt.imshow(image.rotate(-90))

plt.show()
```

```python
# crop image

# load image
image = Image('open_house.jpg')

cropped_image = image.crop((300,300,1000,1000))

cropped_image.show()
```
