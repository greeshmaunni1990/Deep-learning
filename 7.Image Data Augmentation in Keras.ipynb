```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

## Horizontal Shift

```python
# Load image
image = load_img('bird.jpg')

# convert image to numpy array
image_array = img_to_array(image)

# expand the dimension (adding a dimension)
samples = np.expand_dims(image_array, 0)

# create data generator
datagen = ImageDataGenerator(width_shift_range=[-100, 100])

#prepare iterator
iterator = datagen.flow(samples, batch_size=1)

# for samples and plot
for i in range(9):
    plt.subplot(330 + 1 + i)
    
    # generate batch of images
    batch = iterator.next()
    
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

## Brightness

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# load image
image = load_img('bird.jpg')

# convert to arrays
image_to_array = img_to_array(image)

# extend the dimension
samples = np.expand_dims(image_to_array, 0)

# create data generator
# value less than 1.0 will darken and value greater than 1.0 will lighten the image
# 1.0 will have no effect
datagen = ImageDataGenerator(brightness_range=[0.2, 1.0])

# for lighter image
# datagen = ImageDataGenerator(brightness_range=[1.0, 1.5])

# prepare the data for augmentation
iterator = datagen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = iterator.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

## Zoom

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# load image
image = load_img('bird.jpg')

# convert to arrays
image_to_array = img_to_array(image)

# extend the dimension
samples = np.expand_dims(image_to_array, 0)

# create data generator
# value less than 1.0 will zoom in the image(makes it bigger) 
# value greater than 1.0 will zoom out the image(makes it smaller)
# 1.0 will have no effect
# datagen = ImageDataGenerator(zoom_range=[0.5, 1.0])

# zoom out the image
datagen = ImageDataGenerator(zoom_range=[1.5, 1.0])


# for lighter image
# datagen = ImageDataGenerator(brightness_range=[1.0, 1.5])

# prepare the data for augmentation
iterator = datagen.flow(samples, batch_size=1)

for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = iterator.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```
